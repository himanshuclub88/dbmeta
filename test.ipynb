```python
from metadb import FolderDB

db = FolderDB(base_path="genrated_batch_logs", base_metadata="metadata.json")
db.execution_info.show()
db.stats.show()
db.inputs.show()
```

```python
db.sql("""
SELECT iid, start_time, rows_in, rows_out, files_read, status
FROM execution_info
JOIN stats USING(iid)
JOIN inputs USING(iid)
""").show()
```

```python
db.sql("""
    SELECT iid, duration_sec as second
    FROM execution_info
    ORDER BY duration_sec DESC
""").show()
```

```python
db.stats.where(("rows_in", ">", 300000)).all()
```

```python
db.execution_info.where(("status", "=", "FAILED")).show()
```

```python
db.sql("""
SELECT *
FROM execution_info
JOIN stats USING(iid)
JOIN inputs USING(iid)
WHERE status = 'FAILED'
order by start_time
""").show()
```

```python
db.sql("""
SELECT start_time as day,
       AVG(duration_sec)
FROM execution_info
GROUP BY day
""").all()
```
